```python
sri_ventas_2023 = pd.read_csv(r"C:\Users\juanm\OneDrive\Desktop\DAT\sri_ventas_2023.csv", encoding='latin-1', sep= '|')

sri_ventas_2023.head(10)
```

```python
 sri_ventas_2023= sri_ventas_2023.drop_duplicates()
```

```python
sri_ventas_2023.describe()
```

```python

df = pd.DataFrame(sri_ventas_2023)

df['TOTAL_COMPRAS'] = pd.to_numeric(df['TOTAL_COMPRAS'].astype(str).str.replace(',', '.'), errors='coerce')

df['TOTAL_VENTAS'] = pd.to_numeric(df['TOTAL_VENTAS'].astype(str).str.replace(',', '.'), errors='coerce')

df['COMPRAS_RISE'] = pd.to_numeric(df['COMPRAS_RISE'].astype(str).str.replace(',', '.'), errors='coerce')

df['IMPORTACIONES'] = pd.to_numeric(df['IMPORTACIONES'].astype(str).str.replace(',', '.'), errors='coerce')

df['COMPRAS_NETAS_TARIFA_0'] = pd.to_numeric(df['COMPRAS_NETAS_TARIFA_0'].astype(str).str.replace(',', '.'), errors='coerce')

df['COMPRAS_NETAS_TARIFA_12'] = pd.to_numeric(df['COMPRAS_NETAS_TARIFA_12'].astype(str).str.replace(',', '.'), errors='coerce')

df['EXPORTACIONES'] = pd.to_numeric(df['EXPORTACIONES'].astype(str).str.replace(',', '.'), errors='coerce')

df['VENTAS_NETAS_TARIFA_0'] = pd.to_numeric(df['VENTAS_NETAS_TARIFA_0'].astype(str).str.replace(',', '.'), errors='coerce')

df['VENTAS_NETAS_TARIFA_12'] = pd.to_numeric(df['VENTAS_NETAS_TARIFA_12'].astype(str).str.replace(',', '.'), errors='coerce')

df.head(10)
```

```python
print(df.info())

df.describe()
```

```python
df['VENTAS_NETAS_TARIFA_15'] = df['VENTAS_NETAS_TARIFA_12'] * 1.03

df['COMPRAS_NETAS_TARIFA_15'] = df['COMPRAS_NETAS_TARIFA_12'] * 1.03 
df
```

```python
group = df.groupby(['PROVINCIA']).agg({'TOTAL_COMPRAS': ['count', 'mean', 'max']})
group
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Graficar el número de compras por provincia
plt.figure(figsize=(40, 15))
# Graficar la media de compras por provincia
plt.subplot(1, 2, 2)
group['TOTAL_COMPRAS']['mean'].plot(kind='bar', color='lightcoral')
plt.title('Media de Compras por Provincia')
plt.xlabel('Provincia')
plt.ylabel('Media de Compras')
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

```python
sig = df[(df['TOTAL_COMPRAS'] > 1000000) & (df['TOTAL_VENTAS'] > 1000000)]
print(sig.info())
```

```python
group_1 = df.groupby(['PROVINCIA']).agg({'TOTAL_COMPRAS': ['count', 'mean', 'max']})
group_1
```

```python
group_2 = sig.groupby(['PROVINCIA']).agg({'TOTAL_COMPRAS': ['count', 'mean', 'max']})
group_2
```

```python
dif = (group_2 / group_1) *100
dif
```

```python
# Graficar la diferencia de compras por provincia
plt.figure(figsize=(45, 15))
# Graficar la media de compras por provincia
plt.subplot(1, 2, 2)
dif['TOTAL_COMPRAS']['count'].plot(kind='bar', color='blue')
plt.title('% COMPRAS SIGNIFICATIVAS')
plt.xlabel('Provincia')
plt.ylabel('COUNT')
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```
